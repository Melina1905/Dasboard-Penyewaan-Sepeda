## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Pengaruh Musim Terhadap Jumlah Penyewaan Sepeda?

```python
#Mengelompokkan data berdasarkan musim
season_rentals = day_df.groupby(by="season").agg({
    "cnt": ["sum", "mean", "max"],
    "casual": ["sum", "mean"],
    "registered": ["sum", "mean"]
}).reset_index()

#variabel untuk plotting
x_labels= ["Spring", "Summer", "Fall", "Winter"]
x_values= season_rentals ["season"]
y_total= season_rentals[("cnt","sum")]
y_casual= season_rentals[("casual","sum")]
y_registered= season_rentals[("registered","sum")]

#plot
plt.figure(figsize=(10,5))
plt.plot(x_values, y_total, marker= 'o', linewidth=2, color="#72BCD4", label= "Total Peminjaman")
plt.plot(x_values, y_casual, marker= 'o', linewidth=2, color="#FF7F50", label= "Casual")
plt.plot(x_values, y_registered, marker= 'o', linewidth=2, color="#4682B4", label= "Registered")

plt.title("Total Peminjaman Sepeda Berdasarkan Musim", loc="center", fontsize=15)
plt.xlabel("Musim", fontsize=12)
plt.ylabel("Total Peminjaman Sepeda", fontsize=12)
plt.xticks(ticks=x_values, labels=x_labels, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

### Pertanyaan 2:

```python
#mengelompokkan data berdasarkan jam dan hari kerja/hari libur
hour_rentals=hour_df.groupby(["hr", "workingday"]).agg({
    "cnt": ["sum"],
    "casual": ["sum"],
    "registered": ["sum"]
}).reset_index()

#Perbaiki nama kolom multi-index agar mudah dipanggil
hour_rentals.columns = ['hr', 'workingday', 'cnt_sum', 'casual_sum', 'registered_sum']

#pisahkan berdasarkan hari kerja dan hari libur
workingday_df=hour_rentals[hour_rentals["workingday"]==1]
non_workingday_df=hour_rentals[hour_rentals["workingday"]==0]

overall_rentals = hour_rentals.groupby("hr").agg({
    "cnt_sum": "sum"
}).reset_index()

#variabel untuk plotting
x_values= workingday_df["hr"]
y_working_total= workingday_df[("cnt_sum")]
y_nonworking_total= non_workingday_df[("cnt_sum")]
y_overall = overall_rentals["cnt_sum"]

#plot
plt.figure(figsize=(10,5))
plt.plot(x_values, y_working_total, marker='o', linewidth=2, color="#72BCD4",label="Hari Kerja")
plt.plot(x_values, y_nonworking_total, marker='o', linewidth=2, color="#FF7F50",label="Hari Libur")
plt.plot(x_values, y_overall, marker='o', linewidth=2, linestyle="--", color="#2E8B57", label="Tren Jam (Keseluruhan)")

plt.title("Tren Penyewaan Sepeda Berdasarkan Jam (Hari Kerja vs Hari Libur)", loc="center", fontsize=15)
plt.xlabel("Jam", fontsize=12)
plt.ylabel("Total Peminjaman Sepeda", fontsize=12)
plt.xticks(ticks=x_values, fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

**Insight:**
- Pertanyaan 1: Bagaimana Pengaruh Musim Terhadap Jumlah Penyewaan Sepeda?
  - Jumlah penyewa sepeda paling banyak terjadi pada musim panas dan gugur, sedangkan yang paling rendah ada di musim semi dan gugur
  - Ini menggambarkan bahwa peminjaman sepeda dipengaruhi oleh musim, cuaca hangat mendukung aktivitas diluar ruangan
- Pertanyaan 2: Bagaimana tren penyewaan sepeda berdasarkan jam dalam sehari?
  - Jumlah penyewa sepeda paling banyak berkisar antara jam berangkat kerja yaitu jam 08.00 dan jam pulang kerja yaitu 17.00-18.00
  - Jumlah penyewa sepeda paling banyak terjadi di hari kerja
  - Dari kedua hal diatas dan menurut grafik diatas, kemungkinan besar penyewa sepeda adalah para pekerja

## Analisis Lanjutan (Opsional)

```python
#mengelompokkan data berdasarkan pengguna terdaftar
rfm_df=hour_df.groupby(by="registered", as_index=False).agg({
    "dteday": "max",
    "instant": "count",
    "cnt": "sum"
})

#mengubah nama kolom agar sesuai dengan RFM
rfm_df.columns=["user_id", "max_order_timestamp","frequency","monetary"]

#menghitung berapa hari sejak transaksi terakhir
rfm_df["max_order_timestamp"]=pd.to_datetime(rfm_df["max_order_timestamp"]).dt.date
recent_date = hour_df["dteday"].max().date()
rfm_df["recency"]=rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

#menghapus kolom tanggal terakhir setelah digunakan untuk recency
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

rfm_df.head()
```

```python
#pengelompokkan jumlah sepeda yang disewa per hari
daily_trend= hour_df.groupby("dteday")["cnt"].sum().reset_index()

#plot tren harian
plt.figure(figsize=(15,6))
sns.lineplot(x="dteday", y="cnt", data=daily_trend, marker="o", color="#72BCD4")

plt.title("Tren Pengguna Sepeda per Hari", fontsize=16)
plt.xlabel("Tanggal",fontsize=14)
plt.ylabel("Jumlah Sepeda Disewa", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
                                 
```

**Insight:**
- Tren menunjukkan adanya peningkatan jumlah penyewaan sepeda dari waktu ke waktu
- Di bagian akhir grafik terlihat penurunan jumlah penyewaan, yang bisa berarti beberapa pelanggan mulai jarang menggunakan layanan tersebut
